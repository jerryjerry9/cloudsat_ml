```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
# For Chinese font
from matplotlib.font_manager import FontProperties
font = FontProperties(fname="NotoSansCJKtc-Regular.otf", size=10)
from sklearn.cluster import KMeans
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import SpectralClustering
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('zjd20150611.csv')
```

```python
print(data.head)
```

```python
cloud2015=np.empty((3951681,17))
cloud2015[:,:]=data.iloc[0:3951681,:]
cloudsize=cloud2015[:,5]
print(cloud2015.shape)
print(cloudsize)
```

```python
#0 granual  1 julian 2 cloud num 3 lat 4 lon 5 size 6 top  7 bot 8 landsea 
#9  0dbz mask  10 0dbz h   11 10dbz mask  12 10dbz h 13edge 14 edge 15 hori  16 system
aaaa=cloud2015[np.where((cloud2015[:,4]>70) & (cloud2015[:,4]<150)&(cloud2015[:,3]>-25) & (cloud2015[:,3]<25)&(cloud2015[:,6]>6000) & (cloud2015[:,7]<2000) ),:]
#cloud2015((cloud2015[:,5] >1))
print(aaaa.shape)
```

```python
nor_size = preprocessing.scale(aaaa[0,:,5],axis=0)
nor_top = preprocessing.scale(aaaa[0,:,6],axis=0)
nor_bot = preprocessing.scale(aaaa[0,:,7],axis=0)
nor_0dbz = aaaa[0,:,9]
nor_10dbz = aaaa[0,:,11]
nor_hori = preprocessing.scale(aaaa[0,:,15],axis=0)
print(nor_size.shape)
```

```python
#print(nor_top)
testsample=np.empty((7170,6))
testsample2=np.empty((6,7170))
testsample[:,0]=nor_top[:,]
testsample[:,1]=nor_bot[:,]
testsample[:,2]=nor_size[:,]
testsample[:,3]=nor_0dbz[:,]
testsample[:,4]=nor_10dbz[:,]
testsample[:,5]=nor_hori[:,]
testsample2[0,:]=nor_top[:,]
testsample2[1,:]=nor_bot[:,]
testsample2[2,:]=nor_size[:,]
testsample2[3,:]=nor_0dbz[:,]
testsample2[4,:]=nor_10dbz[:,]
testsample2[5,:]=nor_hori[:,]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
linked = linkage(testsample[0:200,:], 'single')
labelList = range(0,200)

plt.figure(figsize=(20, 12))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=20, affinity='euclidean', linkage='ward')  
hc.fit_predict(testsample2[:,:].T) 

print(hc.labels_)
#print(hc.labels_==1)
```
